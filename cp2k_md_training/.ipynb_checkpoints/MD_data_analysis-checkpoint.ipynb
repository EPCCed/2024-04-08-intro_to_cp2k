## _Ab initio_ molecular dynamics data

### Check energy conversation

````python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
def retrieve_ener(ener_file):
    """Converts CP2K *-1.ener file into pandas.DataFrame"""
    header = re.split(r"\s{2,}", open(ener_file).readline().strip("\n"))[1:]
    data = np.loadtxt(ener_file, skiprows = 1)
    ener_data = pd.DataFrame(
        columns = header,
        data = data
    )
    return ener_data
````

````python
ener_file = 'ref_data/npt-md-600K/sys-1.ener'
ener_data = retrieve_ener(ener_file)
ener_data.head()
````

````python
plt.figure(figsize = (3.2, 3.0))
plt.plot(ener_data['Time[fs]'], ener_data['Cons Qty[a.u.]'])
plt.xlabel('Time [fs]')
plt.ylabel('Cons Qty [a.u.]')
# plt.savefig('temperature.png', dpi = 600)
plt.show()
````

````python
plt.figure(figsize = (3.2, 3.0))
plt.plot(ener_data['Time[fs]'], ener_data['Cons Qty[a.u.]'], label = "Cons Qty")
plt.plot(ener_data['Time[fs]'], ener_data['Pot.[a.u.]'], label = "Pot E")
plt.xlabel('Time [fs]')
plt.ylabel('Energy [a.u.]')
plt.legend()
# plt.savefig('temperature.png', dpi = 600)
plt.show()
````

If the constant of motion is not well conserved, try to

*   Make `EPS_SCF` tighter (to reduce drift)
*   Make `TIMESTEP` shorter (to reduce fluctuations)
*   Play with `EXTRAPOLATION_ORDER` (to reduce drift and or instabilities)

### System equilibration check

To judge if a system is well equilibrated is not easy. At least the temperature and the potential energy of the system must oscillate around an average and be free of long term drift.

As a rule of thumb, discard 1/3 of the trajectory, use 2/3 for data analysis.

````python
target_temp = 600
plt.figure(figsize = (3.2, 3.0))
plt.plot(ener_data['Time[fs]'], ener_data['Temp[K]'])
plt.axhline(y = target_temp, color = "k", ls = "-.")
plt.xlabel('Time [fs]')
plt.ylabel('Temperature [K]')
# plt.savefig('temperature.png', dpi = 600)
plt.show()
````

### Radial distribution function

#### 1. Set cell and pbc for each frame in the trajectory

CP2K does not write cell parameters into `*-1.xyz` trajectory, we have to set the cell parameters when calculate $g(r)$

````python
# Read cell parameters
use_col = list(range(2,11)) # colomns 2 - 11 of *-1.cell file are cell parameters
cell_params = np.loadtxt('ref_data/npt-md-600K/sys-1.cell', skiprows = 1, usecols = use_col)
````

````python
# ensure the length of trajectory is the same with cell paramters
print(len(traj) == len(cell_params))
````

````python
cell_params[0].reshape(3, -1) # reshape the numpy.array into 3 x 3
````

````python
for frame, cell in zip(traj, cell_params):
    frame.set_cell(cell.reshape(3, -1))
    frame.set_pbc((1,1,1)) # frames have to be periodic in order to calculate g(r)
````

#### 2. Check if cell parameters and periodic boundary conditions are correctly set

````python
traj[0].cell
````

````python
traj[0].pbc
````

#### 3. Calculate $g(r)$ 

````python
from ase.geometry.analysis import Analysis
````

````python
images = Analysis(traj[1000:]) # skip the first 1000 frames
````

````python
g_r = images.get_rdf(rmax = 5.0,
                     nbins = 100,
                     elements = ["Mg"],
                     return_dists = True)
````

````python
g_r = np.asarray(g_r)
````

````python
g_r_avg, dists = g_r.mean(axis = 0)
````

````python
g_r_avg
````

````python
fig, ax = plt.subplots(figsize = (3.2, 3.0), constrained_layout = True)
ax.plot(dists, g_r_avg)
ax.set_xlabel(r"$r\ [\mathrm{\AA}]$")
ax.set_ylabel("$g(r)$")
plt.show()
````

### Calculate self diffusion coefficent of liquid Mg
Diffusion Coefficient can be calculated by using the Einstein Equation:

$$
\langle [\boldsymbol{r}(t) - \boldsymbol{r}_0]^2 \rangle = 2nDt
$$

where $\boldsymbol{r}(t)$ and $\boldsymbol{r}_0$ are atomic the positions at time $t$ and 0, $n$ is the degrees of freedom and $D$ is the diffusion coefficient

ASE solves this by linear fitting $\frac{1}{2n} \langle[\boldsymbol{r}(t) - \boldsymbol{r}_0]^2 \rangle = Dt$ with $\mathrm{slope} = D$ 

````python
from ase.md import analysis
````

````python
traj_1200K = read("ref_data/npt-md-1200K/sys-pos-1.xyz", index = ":", format = "xyz")
````

````python
# For an MD simulation with timestep of N, and images written every M iterations, our timestep here is N * M
# we used timestep of N = 0.5 fs, and wrote every M = 10 steps, therefore the timestep = 0.5 * 10 = 5
diffusion_coeff = analysis.DiffusionCoefficient(traj_1200K, timestep = 5)
````

````python
diffusion_coeff.calculate(ignore_n_images = 1000) # skip the first 1000 frames
````

````python
diff_coeff, std = diffusion_coeff.get_diffusion_coefficients()
````

````python
diffusion_coeff.plot()
````

Use the `print_data` method to print the results out.

````python
diffusion_coeff.print_data()
````
